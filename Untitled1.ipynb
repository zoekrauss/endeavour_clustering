```python
from obspy.clients.fdsn import Client
import obspy
import pandas as pd
import scipy.io as sio
import numpy as np
client = Client("IRIS")
```

```python
t1 = obspy.UTCDateTime(2006,7,1)
t2 = obspy.UTCDateTime(2016,6,1)

cat = client.get_events(starttime=t1, endtime=t2, minlatitude=46.779, maxlatitude=48.721, minlongitude=-129.72, maxlongitude=-128.5)
```

```python
authors = [e.origins[0].creation_info.author for e in cat]
print(np.unique(authors))
```

```python
times = [ev.origins[0].time.datetime for ev in cat]
lats = [ev.origins[0].latitude for ev in cat]
lons = [ev.origins[0].longitude for ev in cat]
depths = [ev.origins[0].depth for ev in cat]
mags = [ev.magnitudes[0].mag for ev in cat]
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
fig,ax = plt.subplots(1,figsize=[15,10])
s = ax.scatter(times,lats,c=(mags))
ax.axhline(48.3)
ax.axhline(47.65)
plt.colorbar(s,label='Magnitudes');
```

```python
origin_dict = {'time':times,'latitude':lats,'longitude':lons,'depth':depths,'magnitude':mags,'depths':depths}
```

```python
events = pd.DataFrame.from_dict(origin_dict)
events['time'] = [str(t) for t in events['time']]
events.head()
```

```python
# data dictionary
OutData = {} 

# convert DF to dictionary before loading to your dictionary
OutData['Obj'] = events.to_dict('list')

sio.savemat('osc_national_eqs.mat',OutData)
```
